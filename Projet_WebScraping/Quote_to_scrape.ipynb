## Application Webscraping pour extraire les citations et leurs auteurs

```python
html_quotes_to_scrape = response.text
soup = BeautifulSoup(html_quotes_to_scrape, "html.parser")
soup
```

```python
soup_author = soup.find_all(class_="author")
soup_quotes = soup.find_all(class_="text")
```

```python
liste_author = [x.get_text() for x in soup_author]
liste_text = [y.get_text() for y in soup_quotes]
```

```python
df_quote_by_author = pd.DataFrame({'auteur':liste_author,
                                   'citation':liste_text})
df_quote_by_author
```

```python
# liste global
liste_author_all = []
liste_quote_all = []
liste_next_all = []

# url de base
url_core = "https://quotes.toscrape.com/page/"

# itérateur pour chaque page
no_page = 1 # itérateur

# Boucle 
while no_page != 0:

    # url incrémentée
    url_specific = f"{url_core}{no_page}"
    print(url_specific)

    # request
    response_specific = requests.get(url_specific)

    # parser
    soup_specific = BeautifulSoup(response_specific.text, "html.parser")

    # sélection des auteurs, citations et next
    soup_authors_specific = soup_specific.find_all(class_="author")
    soup_quotes_specific = soup_specific.find_all(class_="text")
    soup_next_specific = soup_specific.find_all(class_="next")

    # liste des auteurs, citations et 
    liste_author_specific = [x.get_text() for x in soup_authors_specific]
    for x in liste_author_specific:
        liste_author_all.append(x)

    liste_quote_specific = [y.get_text() for y in soup_quotes_specific]
    for y in liste_quote_specific:
            liste_quote_all.append(y)
    
    # controleur de l'itération : next
    try :
        soup_next_specific = soup_specific.find("li", class_="next").get_text()
        print(soup_next_specific)
        liste_next_all.append(soup_next_specific)

        # Code Itératif
        if str(soup_next_specific).find("Next") == 1:
            no_page += 1
    except:
        print("Fin du Scraping")
        no_page = 0
        break
```

```python
df_quote_by_author_all = pd.DataFrame({"auteur": liste_author_all, "citation":liste_quote_all})
df_quote_by_author_all.head(10)
```
